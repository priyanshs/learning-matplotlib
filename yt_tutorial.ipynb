```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import pandas as pd

%reload_ext autoreload
%autoreload 2 
```

```python
x_1 = np.linspace(0, 5, 10)
y_1 = x_1 ** 2
plt.plot(x_1, y_1)
plt.title('First Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x_1, y_1, 'r--')
plt.subplot(1, 2, 2)
plt.plot(x_1, y_1, 'g--')
```

```python
fig_1 = plt.figure(figsize=(8, 6),dpi=100)
axes_1 = fig_1.add_axes([0.1,0.1,0.8,0.8])
axes_1.set_xlabel('X Axis')
axes_1.set_ylabel('Y Axis')
axes_1.set_title('Title')
axes_1.plot(x_1, y_1,label="one")
axes_1.plot(y_1, x_1,label="two")
axes_1.legend(loc=0)


# This is us defining another smaller axes within 
axes_2 = fig_1.add_axes([0.45,0.45,0.3,0.3])
axes_2.set_xlabel('X Axis')
axes_2.set_ylabel('Y Axis')
axes_2.set_title('Title')
axes_2.plot(y_1, x_1)

axes_2.text(0,4, "message")
```

```python
fig_2, axes_2 = plt.subplots(nrows=1, ncols=3,figsize=(8, 4))
plt.tight_layout()
axes_2[0].set_title('Plot 1')
axes_2[0].set_xlabel('X0 Axis')
axes_2[0].set_ylabel('Y0 Axis')
axes_2[0].plot(x_1, y_1, "g*")

axes_2[1].set_title('Plot 2')
axes_2[1].set_xlabel('X1 Axis')
axes_2[1].set_ylabel('Y1 Axis')
axes_2[1].plot(x_1, y_1,"bo")

axes_2[2].set_title('Plot 3')
axes_2[2].set_xlabel('X2 Axis')
axes_2[2].set_ylabel('Y2 Axis')
axes_2[2].plot(x_1, y_1,"r--")
```

'b' as blue
'g' as green
'r' as red
'c' as cyan
'm' as magenta
'y' as yellow
'k' as black
'w' as white
